```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
import windrose
from windrose import WindroseAxes
import matplotlib.cm as cm
```

```python
kunak = pd.read_csv('kunak_filled.csv', parse_dates=[0], index_col=0)
pollutants = kunak.drop(columns=['Dew Point (C)', 'Humidity (%)', 'Temp (C)'])

co = pd.read_csv('co.csv', parse_dates=[0], index_col=0)

kunak.head()
```

```python
fig, ax = plt.subplots(3, figsize=(8,10), sharex=True)
pollutants.plot(ax=ax[0],grid=True, xlabel='')
pollutants.resample('D').mean().plot(ax=ax[1],grid=True, xlabel='')
pollutants.resample('W').mean().plot(ax=ax[2],grid=True, xlabel='')

plt.suptitle('Hourly, Daily and Weekly mean values of pollutants')
plt.tight_layout()
plt.savefig("Plots/Resampling/resampling.png")
plt.show()
```

```python
pollutants_24 = pollutants.groupby(kunak.index.hour).mean()
pollutants_week = pollutants.groupby(kunak.index.dayofweek).mean()
pollutants_week.loc[:,'O3 (ppb)'] = pollutants_week['O3 (ppb)']/3
pollutants_week.loc[:,'NO2 (ppb)'] = pollutants_week['NO2 (ppb)']/4
pollutants_week.rename(columns={'O3 (ppb)':'O3 (ppb/3)','NO2 (ppb)':'NO2 (ppb/4)'}, inplace=True)
```

```python
fig, ax = plt.subplots(2, figsize=(8,8))

pollutants_24.plot(xlabel='Hour of the Day', grid=True, ax=ax[0])
pollutants_week.plot(xlabel='Day of the Week', grid=True, ax=ax[1]).legend(loc=2)

ax[0].set_title('Typical 24 hour profile')
ax[0].set_xticks(np.arange(0,24,3))

ax[1].set_title('Typical Weekly profile')
ax[1].set_xticks(range(7))
ax[1].set_xticklabels(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])

plt.tight_layout()
plt.savefig("Plots/Typical/.typical.png")
plt.show()
```

A combination of boxplot and kernel density estimate.

A violin plot plays a similar role as a box and whisker plot. It shows the
distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared. Unlike
a box plot, in which all of the plot components correspond to actual
datapoints, the violin plot features a kernel density estimation of the
underlying distribution. (Να εξηγήσω την αξία τους)

### Decomposition

```python
daily = kunak.resample('D').mean()
```

```python
h2s = seasonal_decompose(daily['H2S (ppb)'])
no2 = seasonal_decompose(daily['NO2 (ppb)'])
o3 = seasonal_decompose(daily['O3 (ppb)'])
pm25 = seasonal_decompose(daily['PM2.5 (ug/m3)'])
```

```python
pollutants_list = [h2s, no2, o3, pm25]
for p in pollutants_list:
    fig,ax = plt.subplots(4, figsize=(8,6), sharex=True)
    p.observed.plot(rot=45, ax=ax[0])
    ax[0].set_ylabel('Observed')
    p.trend.plot(rot=45, ax=ax[1])
    ax[1].set_ylabel('Trend')
    p.seasonal.plot(rot=45, ax=ax[2])
    ax[2].set_ylabel('Seasonal')
    p.resid.plot(linestyle=' ', marker='o', rot=45, ax=ax[3])
    ax[3].set_ylabel('Residual')
    
    plt.xlabel("")
    plt.suptitle(f"Components of {p.observed.name}")
    figname = 'Plots/Decomposition/' + p.observed.name[:2] + '.png'
    plt.savefig(figname)
```

#### Περιοδικότητα

```python
days = 8
plt.figure(figsize=(10,4))

h2s.seasonal[:days].plot(grid=True, label='H2S')
no2.seasonal[:days].plot(grid=True, label='NO2')
pm25.seasonal[:days].plot(grid=True, label='PM2.5')
o3.seasonal[:days].plot(grid=True, label='O3')

plt.xticks(daily.index[:days], daily.index.day_name()[:days], rotation=45)
plt.legend(loc='lower left')
plt.xlabel('')
plt.yticks([])
plt.tight_layout()
plt.savefig('Plots/Decomposition/periodicity.png')
plt.show()
```

### Wind effect

```python
kunak_wind = pd.read_csv('kunak_wind.csv')
direction_order = ["N","NE","E","SE","S","SW","W","NW","CALM"]
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,8))
sns.barplot(x='wind direction',y= 'H2S (ppb)', data=kunak_wind, order=direction_order, color='yellowgreen', ax=axes[0,0])
sns.barplot(x='wind direction',y= 'NO2 (ppb)', data=kunak_wind, order=direction_order, color="salmon", ax=axes[0,1])
sns.barplot(x='wind direction',y= 'O3 (ppb)', data=kunak_wind, order=direction_order, color="skyblue", ax=axes[1,0])
sns.barplot(x='wind direction',y= 'PM2.5 (ug/m3)', data=kunak_wind, order=direction_order, color="slategrey", ax=axes[1,1])

fig.suptitle("Wind Effect", y=0.94)
#fig.tight_layout()
fig.savefig('Plots/Wind/barplots.png', bbox_inches='tight')
plt.show()
```

### Windspeed > 0.5 m/s

```python
calmless = kunak_wind[kunak_wind.loc[:,'windspeed']>0.5]
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(2,2,1, projection="windrose")
ax.bar(calmless['direction'], calmless['H2S (ppb)'], nsector=8, bins=[0,10,20,40], cmap=cm.Greens)
ax.set_title('H2S (ppb)')
ax.legend(loc=2)

ax = fig.add_subplot(2,2,2, projection="windrose")
ax.bar(calmless['direction'], calmless['NO2 (ppb)'], nsector=8, bins=[0,20,50,100], cmap=cm.Reds)
ax.set_title('NO2 (ppb)')
ax.legend(loc=2)

ax = fig.add_subplot(2,2,3, projection="windrose")
ax.bar(calmless['direction'], calmless['O3 (ppb)'], nsector=8, bins=[0,20,60,80], cmap=cm.Blues)
ax.set_title('O3 (ppb)')
ax.legend(loc=2)

ax = fig.add_subplot(2,2,4, projection="windrose")
ax.bar(calmless['direction'], calmless['PM2.5 (ug/m3)'], nsector=8, bins=[0,5,10,25], cmap=cm.Greys)
ax.set_title('PM2.5 (ug/m3)')
ax.legend(loc=2)

plt.suptitle("Pollutants Windroses")
plt.tight_layout()
plt.savefig("Plots/Wind/windrose.png")
plt.show()
```
