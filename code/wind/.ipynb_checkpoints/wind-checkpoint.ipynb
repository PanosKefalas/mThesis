```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
import windrose
```

```python
may = pd.read_csv('May.csv')
june = pd.read_csv('June.csv')
july = pd.read_csv('July.csv')
```

```python
wind = pd.concat([may[135:],june,july])
```

```python
wind.index = pd.to_datetime(wind[['year', 'month', 'day', 'hour', 'minute']])
wind_clean = wind[['windspeed', 'direction']].resample('H').mean().round(2)
wind_clean
```

```python
wind_droped = wind_clean.dropna(how='any',axis=0)
wind_droped.to_csv('wind.csv')
```

```python
bins = bins = np.linspace(0,360,9)+22.5
bins = np.insert(bins,0,0)
labels = ["N","NE","E","SE","S","SW","W","NW","N2"]
```

```python
wind_droped['wind direction'] = pd.cut(wind_droped.direction, bins=bins, labels=labels, include_lowest=True)
wind_droped['wind direction'].replace('N2','N', inplace=True)
wind_droped.head()
```

```python
freq = wind_droped['wind direction'].value_counts()
wind_droped.loc[:,'frequency'] = freq[wind_droped['wind direction']].values / freq.sum()
wind_droped
```

```python
windrose.wrcontourf(wind_droped['direction'], wind_droped['windspeed'])
plt.show()
```
