```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error, r2_score, mean_squared_error
plt.style.use('seaborn-poster')
```

```python
df_raw = pd.read_excel('aqmesh-final-raw.xlsx', parse_dates=[0], index_col=0)
df_raw.index.name = 'Datetime'
df = df_raw[['NO2 PreScaled ppb', 'O3 PreScaled ppb', 'Humidity %RH', 'Temperature C']]
df.columns = (['NO2 (ppb)', 'O3 (ppb)', 'Humidity (%)', 'Temp (C)'])
df
```

```python
df.isna().sum()
```

### Testing

```python
target = 'Temp (C)'
target_name = target
df_test = df.copy()
df_dummy_filled = df_test.fillna(method='bfill')
nan_list = []
breaks = 30
missing_values = 12
for b in range(breaks+1):
    nan_list.append(int(df_test.shape[0]/breaks*b))

for i in nan_list[1:]:
    df_test[target].iloc[i:i+missing_values] = np.nan

df_test.isna().sum()
```

```python
plt.figure(figsize=(14,6))
df_test[target][nan_list[2]-50:nan_list[2]+50].plot()
plt.ylabel(f"{target}")
plt.title(f"Unfilled {target} testing sample")
plt.tight_layout()
plt.savefig(f'Plots/imputation/{target_name}_missing.png')
plt.show()
```

#### KNN Imputer

```python
imputer = KNNImputer(n_neighbors=5)
df_knn = df_test.copy()
df_knn[:] = imputer.fit_transform(df_test)
```

```python
errors_knn_list = [mean_absolute_error(df_knn[target], df_dummy_filled[target]),
               mean_squared_error(df_knn[target], df_dummy_filled[target]),
              r2_score(df_knn[target], df_dummy_filled[target])]

errors_knn = pd.Series(errors_knn_list,  index=['MAE','MSE','R2'])
```

#### Interpolation

```python
df_linear = df_test.copy()
df_linear.interpolate(method='linear', inplace=True)
```

```python
errors_linear_list = [mean_absolute_error(df_linear[target], df_dummy_filled[target]),
               mean_squared_error(df_linear[target], df_dummy_filled[target]),
              r2_score(df_linear[target], df_dummy_filled[target])]

errors_linear = pd.Series(errors_linear_list,  index=['MAE','MSE','R2'])
```

```python
df_cubic = df_test.copy()
df_cubic.interpolate(method='cubic', order=3,inplace=True)
```

```python
errors_cubic_list = [mean_absolute_error(df_cubic[target], df_dummy_filled[target]),
               mean_squared_error(df_cubic[target], df_dummy_filled[target]),
              r2_score(df_cubic[target], df_dummy_filled[target])]

errors_cubic = pd.Series(errors_cubic_list,  index=['MAE','MSE','R2'])
```

#### Plotting

```python
plt.figure(figsize=(14,6))

df[target][nan_list[2]-50:nan_list[2]+50].plot(label=f'Actual {target}')
df_linear[target][nan_list[2]-50:nan_list[2]+50].plot(linestyle='--', label=f'Linear Interpolation {target}')
df_cubic[target][nan_list[2]-50:nan_list[2]+50].plot(linestyle='--', label=f'Cubic Interpolation {target}')
df_knn[target][nan_list[2]-50:nan_list[2]+50].plot(linestyle='--', label=f'KNN Imputation {target}')

plt.title(f"Filling {target} Missing Values")
plt.legend()
plt.ylabel(f"{target}")
plt.tight_layout()
plt.savefig(f'Plots/imputation/{target}_Imputation.png')
plt.show()
```

### Errors

```python
errors = pd.concat([errors_knn, errors_linear, errors_cubic], axis=1)
errors.columns = ['KNN', 'Linear', 'Cubic']
errors = errors.round(3).T
errors
```

```python
errors.to_csv(f"{target}_errors.csv")
```
