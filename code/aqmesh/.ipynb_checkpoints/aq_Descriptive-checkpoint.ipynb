```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
import windrose as wr
import matplotlib.cm as cm
```

```python
aq = pd.read_csv('aq_filled.csv', parse_dates=[0], index_col=0)
pollutants = aq.iloc[:,[0,1]]
aq
```

```python
fig, ax = plt.subplots(3, figsize=(8,10), sharex=True)
pollutants.plot(ax=ax[0],grid=True, xlabel='')
pollutants.resample('D').mean().plot(ax=ax[1],grid=True, xlabel='')
pollutants.resample('W').mean().plot(ax=ax[2],grid=True, xlabel='')

plt.suptitle('Hourly, Daily and Weekly mean values of pollutants')
plt.tight_layout()
plt.savefig("Plots/Resampling/resampling.png")
plt.show()
```

```python
aq_24 = aq.groupby(aq.index.hour).mean()
aq_week = aq.groupby(aq.index.dayofweek).mean()
```

```python
fig, ax = plt.subplots(2, figsize=(8,8))

aq_24.plot(xlabel='Hour of the Day', grid=True, ax=ax[0])
aq_week.plot(xlabel='Day of the Week', grid=True, ax=ax[1])

ax[0].set_title('Typical 24 hour profile')
ax[0].set_xticks(np.arange(0,24,3))

ax[1].set_title('Typical Weekly profile')
ax[1].set_xticks(range(7))
ax[1].set_xticklabels(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])

plt.tight_layout()
plt.savefig("Plots/Typical/typical.png")
plt.show()
```

### Decomposition

```python
daily = aq.resample('D').mean()
```

```python
no2 = seasonal_decompose(daily['NO2 (ppb)'], period=7)
o3 = seasonal_decompose(daily['O3 (ppb)'], period=7)
```

```python
pollutants_list = [no2, o3]
for p in pollutants_list:
    fig,ax = plt.subplots(4, figsize=(8,6), sharex=True)
    p.observed.plot(rot=45, ax=ax[0])
    ax[0].set_ylabel('Observed')
    p.trend.plot(rot=45, ax=ax[1])
    ax[1].set_ylabel('Trend')
    p.seasonal.plot(rot=45, ax=ax[2])
    ax[2].set_ylabel('Seasonal')
    p.resid.plot(linestyle=' ', marker='o', rot=45, ax=ax[3])
    ax[3].set_ylabel('Residual')
    
    plt.xlabel("")
    plt.suptitle(f"Components of {p.observed.name}")
    figname = 'Plots/Decomposition/' + p.observed.name[:2] + '.png'
    plt.savefig(figname)
```

#### Περιοδικότητα

```python
days = 8
plt.figure(figsize=(10,4))

no2.seasonal[:days].plot(grid=True, label='NO2')
o3.seasonal[:days].plot(grid=True, label='O3')

plt.xticks(daily.index[:days], daily.index.day_name()[:days], rotation=45)
plt.legend(loc=2)
plt.xlabel('')
plt.savefig("Plots/Decomposition/periodicity.png")
plt.show()
```

### Wind Effect

```python
aq_wind = pd.read_csv('aq_wind.csv')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,4))
sns.barplot(x='wind direction',y= 'NO2 (ppb)', data=aq_wind, order=["N","NE","E","SE","S","SW","W","NW","CALM"], color='salmon', ax=axes[0]).set(xlabel=None)
sns.barplot(x='wind direction',y= 'O3 (ppb)', data=aq_wind, order=["N","NE","E","SE","S","SW","W","NW","CALM"], color="skyblue", ax=axes[1])
plt.tight_layout()
plt.savefig("Plots/Wind/wind_effect.png")
plt.show()
```
