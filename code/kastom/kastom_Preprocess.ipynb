```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv('KASTOM2022.csv', parse_dates=[0], index_col=0)
df_raw = df_raw.loc['2022-04-10':]
df = df_raw[['temperature', 'humidity','pm_10', 'pm_2_5', 'pm_1_0', 'no2_ppb','co_μg/m3']]
df.head()
```

```python
df.columns = ['Temp (C)', 'Humidity (%)', 'PM10 (ug/m3)', 'PM2.5 (ug/m3)', 'PM1 (ug/m3)', 'NO2 (ppb)', 'CO (μg/m3 10²)']
df.loc[:,'CO (μg/m3 10²)'] = df['CO (μg/m3 10²)'] /10
df.head()
```

```python
print(df.isna().sum())
print((df<0).sum())
```

```python
df[df<0] = np.nan
print(df.isna().sum())
print((df<0).sum())
```

```python
imputer = KNNImputer(n_neighbors=5, weights='distance')
df[:]= imputer.fit_transform(df).round(2)
df.isna().sum()
```
