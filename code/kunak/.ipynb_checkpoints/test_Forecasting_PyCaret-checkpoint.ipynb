```python
from pycaret.regression import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv('kunak_filled.csv', parse_dates=[0], index_col=0).drop(columns='Pressure (hPa)')
df = df_raw.copy()
df.tail()
```

```python
steps = 7*24
train = df[:-steps]
test = df[-steps:]

target = 'H2S (ppb)'
numeric_features = df.columns.drop(target).to_list()
```

```python
s = setup(data=train, test_data=test, target=target, fold_strategy='timeseries',
          numeric_features=numeric_features,
          fold=3,  silent=True, session_id=123, verbose=1)
```

```python
model = create_model('rf')
```

```python
predictions = predict_model(model, data=test)
```

```python
plt.style.use('seaborn-notebook')
predictions[[target, 'Label']].plot(figsize=(14,4))
plt.legend([f'Actual {target}', f'Predicted {target}'])
plt.show()
```
