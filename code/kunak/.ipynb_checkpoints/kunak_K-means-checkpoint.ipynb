```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

### Συσταδοποίηση ημερήσιων μέσων

```python
df = pd.read_csv('kunak_filled.csv', parse_dates=[0], index_col=0).resample('D').mean()
pollutants = df[['H2S (ppb)','NO2 (ppb)', 'O3 (ppb)', 'TSP (ug/m3)']]
```

```python
scaler = MinMaxScaler()
pollutants_scaled = pollutants.copy()
pollutants_scaled[:] = scaler.fit_transform(pollutants)
```

```python
pca = PCA(0.8)
pollutants_pca = pca.fit_transform(pollutants_scaled)
print(pca.n_components_)
```

```python
n_pcs = pca.n_components_
important = [np.abs(pca.components_)[i].argmax() for i in range(n_pcs)]
names = pollutants.columns
important_features = [names[important[i]] for i in range(n_pcs)]
print(important_features)
```

```python
inertias = []
for k in range(2,12):
    kmeans = KMeans(k)
    kmeans.fit(pollutants_pca)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(inertias)
plt.show()
```

```python
kmeans = KMeans(4)
kmeans.fit(pollutants_pca)
```

```python
df_clustered = df.copy()
df_clustered['PC1'] = pollutants_pca[:,0]
df_clustered['PC2'] = pollutants_pca[:,1]
df_clustered['cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_clustered['PC1'], y=df_clustered['PC2'], hue=df_clustered['cluster'], palette="tab10")
plt.show()
```

```python
profiles = round(df_clustered.drop(columns=['PC1','PC2','Pressure (hPa)','Dew Point (C)','Temp (C)','Humidity (%)']),0)
profiles
```

```python
sns.set_style('whitegrid')
sns.set_palette("deep")
fig, ax = plt.subplots(2,2, figsize=(14,10))

sns.boxplot(data=profiles.loc[profiles['cluster']==0].drop(columns='cluster'), ax=ax[0,0], saturation=.5)
sns.boxplot(data=profiles.loc[profiles['cluster']==1].drop(columns='cluster'), ax=ax[0,1], saturation=.5)
sns.boxplot(data=profiles.loc[profiles['cluster']==2].drop(columns='cluster'), ax=ax[1,0], saturation=.5)
sns.boxplot(data=profiles.loc[profiles['cluster']==3].drop(columns='cluster'), ax=ax[1,1], saturation=.5)


ax[0,0].set_title('Cluster 0')
ax[0,1].set_title('Cluster 1')
ax[1,0].set_title('Cluster 2')
ax[1,1].set_title('Cluster 3')
plt.show()
```

```python
sns.set_palette("deep")
df_clustered['cluster'].value_counts().plot(kind='pie',autopct='%1.1f%%', figsize=(5,5))
plt.title('Cluster Frequency')
plt.ylabel("")
plt.show()
```
