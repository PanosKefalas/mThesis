# Μελέτη περιοδικότητας με την μέθοδο σειριακής συσχέτισης (Autocorrelation)

```python
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kunak_filled.csv', parse_dates=['Datetime'], index_col='Datetime')
df = df.resample('H').mean()
h2s = df['H2S (ppb)']
h2s.head()
```

#### Επιλογή πλήθους χρονικών υστερήσεων και εφαρμοφή της συνάρτησης αυτοσυσχέτισης

```python
lags=100
fig, ax = plt.subplots(figsize=(10,6))
plot_acf(h2s, lags=lags, ax=ax)
plt.xticks(np.arange(0,lags,24))
plt.grid()
plt.show()
```

***Εμφανής 24ωρη περιοδικότητα με διάστημα εμπιστοσύνης > 95% (σκιασμένη περιοχή)***
